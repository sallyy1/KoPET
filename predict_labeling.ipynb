```python
# 1. 법제처
# unlabeled_path = './laws/unlabeled.jsonl'
# logits_path = './Law_Output/g2/unlabeled_logits.txt'

# 2. CS sharing
unlabeled_path = './cssharing_new/unlabeled.jsonl'
logits_path = './CSsharing_Output_new/g1/unlabeled_logits.txt'

import json

all_data = []

with open(unlabeled_path, 'r', encoding='utf-8') as f:
    for line in f:
        example_json = json.loads(line) # json.load: JSON 형태의 문자열을 dictionary 형태의 자료형으로 반환

        idx = example_json['idx']
        label = str(example_json['label']) if 'label' in example_json else None
        text_a = example_json['text']
        
        print(example_json['idx'], example_json['text'], label)
        print('*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*')
        all_data.append([example_json['idx'], example_json['text'], label])
```

```python
import csv
import numpy as np

idx = -1

with open(logits_path, 'r', encoding='utf-8') as f2:
    strings = f2.readlines()
    for line in strings:
        line = line.rstrip()
        
        if line == '-1':
            continue
            
        idx += 1
        print(line, type(line))
        label_0, label_1 = line.split(' ')
        label_0, label_1 = float(label_0), float(label_1)
        result = np.array([label_0, label_1])
        print(result)
        predictions = np.argmax(result, axis=0)
        print(predictions)
        all_data[idx][2] = predictions # 예측한 label값으로 업데이트
        print(all_data[idx])
        print()
```

```python
for line in all_data:
    print(line)
```

```python
new_f = open('./predict_labeled_data.csv', 'w')
wr = csv.writer(new_f)

wr.writerow(['idx', 'text', 'predicted label'])

for line in all_data:
    wr.writerow(line)

new_f.close()
```
